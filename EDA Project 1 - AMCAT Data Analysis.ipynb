## 1.Statistical Non Visual Analysis

Univariate Analysis -> PDF, Histograms, Boxplots, Countplots, etc.. To find the outliers in each numerical column & understand the probability and frequency distribution of each numerical column. Also, to understand the frequency distribution of each categorical Variable/Column

```python
discrete_df = df.select_dtypes(include=['object'])

numerical_df = df.select_dtypes(include=['int64', 'float64'])
```

```python
def discrete_univariate_analysis(discrete_data):
    for col_name in discrete_data:
        print("*"*10, col_name, "*"*10)
        print(discrete_data[col_name].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', discrete_data[col_name].value_counts())
        print()
```

```python
def numerical_univariate_analysis(numerical_data):
    for col_name in numerical_data:
        print("*"*10, col_name, "*"*10)
        print(numerical_data[col_name].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

## V. Research Question

Research Question 1: Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given. 

```python
import pandas as pd

#Assuming your DataFrame is named 'df'
relevant_specializations = [
    'computer engineering', 'computer science & engineering', 
    'electronics and computer engineering', 'information science engineering',
    'computer science and technology', 'computer networking', 
    'biomedical engineering', 'electronics', 'computer science'
]

relevant_designations = [
    'programming analyst', 'software engineer', 'hardware engineer', 'associate engineer'
]

#Filter data based on relevant specialization and designations
filtered_date = df[df['Specialization'].isin(relevant_specializations) & df['Designation'].isin(relevant_designations)]

#Filter data before the clamied data (Jan 18, 2019)
claimed_date = pd.to_datetime('2019-01-18')
data_before_claimed_date = filtered_date[filtered_date['DOJ'] < claimed_date]

#Filter data within the claimed salary range (2.5-3 lakhs)
claimed_salary_range = (250000,300000)
within_claimed_range = data_before_claimed_date[
    (data_before_claimed_date['Salary'] >= claimed_salary_range[0]) & (data_before_claimed_date['Salary'] <= claimed_salary_range[1])
]

#Calculate the percentage of fresh graduates within the claimed salry range
percentage_within_claimed_range = (len(within_claimed_range) / len(data_before_claimed_date)) *100

#Display the results
print(f"Percentage of fresh graduates within the claimed salary range before Jan 18, 2019: {percentage_within_claimed_range:.2f}%")
     
      
```

```python
#Create a histogram
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.hist(data_before_claimed_date['Salary'],bins=30,alpha=0.7,color='green',label='All Data')
plt.hist(within_claimed_range['Salary'],bins=30,alpha=0.7,color='orange',label='Within claimed Range')

#Highlight the claimed salary range
plt.axvline(x=claimed_salary_range[0],color='red',linestyle='--',linewidth=2,label='Claimed Range Start')
plt.axvline(x=claimed_salary_range[1],color='red',linestyle='--',linewidth=2,label='Claimed Range End')

plt.title('Salary Distribution for Specified Job Titles (Before Jan 18, 2019)')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Claim : Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” 

     Percentage of fresh graduates within the claimed salary range before Jan 18, 2019 is only 13.62%. So, this claim is not completely true.
